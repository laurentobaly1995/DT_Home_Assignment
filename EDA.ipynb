```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample_data/train_data.csv')
```

```python
df.describe()['winBid']
```

```python
unique_per_column = []
for col in df.columns:
  unique_per_column.append(len(df[col].unique()))
print(df.columns)
print(unique_per_column)
```

```python
df.groupby('unitDisplayType').mean()[[ 'winBid']].plot(kind = 'bar', figsize= (5,5))
```

```python
a = pd.crosstab(df.unitDisplayType, df.has_won)
a['totalBids'] = a[0.0] + a[1.0]
a['winPercentage'] = (a[1.0] * 100) / a.totalBids
a['losePercentage'] =  100 - a.winPercentage
a = a[['winPercentage', 'losePercentage']]
a.plot(kind='bar', figsize= (5,5))
```
